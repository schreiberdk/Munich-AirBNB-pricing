# Initial Time Series Forecasting

```python
import numpy as np
np.set_printoptions(legacy='1.25')
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

#### Load cleaned data

```python
df = pd.read_csv("../data/listings_cleaned.csv", parse_dates=["last_review"])
```

#### Drop rows without a review date (needed for time index)

```python
df = df.dropna(subset=["last_review"])
df = df[df["last_review"] >= "2016-01-01"]
```

#### Resample: average price per week

```python
ts = df.set_index("last_review").resample("W")["price"].mean().dropna().to_frame()
ts = ts.rename(columns={"price": "y"}).reset_index()
ts = ts.rename(columns={"last_review": "ds"})  # Prophet expects "ds" and "y"
```

#### Plot

```python
plt.figure(figsize=(12, 5))
plt.plot(ts["ds"], ts["y"])
plt.title("Average Airbnb Price per Week in Munich")
plt.xlabel("Date")
plt.ylabel("Average Price (€)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Time series

```python
from prophet import Prophet
```

#### Initialize and fit model

```python
model = Prophet()
model.fit(ts)
```

#### Forecast 12 weeks ahead

```python
future = model.make_future_dataframe(periods=12, freq="W")
forecast = model.predict(future)
```

#### Plot forecast

```python
fig1 = model.plot(forecast)
fig1.suptitle("Airbnb Price Forecast", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Predicted Price (€)")
plt.show()
```

```python
# Optional: plot components (trend + seasonality)
fig2 = model.plot_components(forecast)
plt.show()
```

#### Train/test split: last 12 weeks as test

```python
train = ts[:-12]
test = ts[-12:]
```

#### Fit on train

```python
model_train = Prophet()
model_train.fit(train)
```

#### Predict on test period

```python
future_test = model_train.make_future_dataframe(periods=12, freq="W")
forecast_test = model_train.predict(future_test)
```

#### Extract relevant forecast

```python
pred = forecast_test[["ds", "yhat"]].set_index("ds").loc[test["ds"]]
```

#### Compare with actuals

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(test["y"], pred["yhat"])
rmse = np.sqrt(mean_squared_error(test["y"], pred["yhat"]))

print(f"MAE: {mae:.2f} €")
print(f"RMSE: {rmse:.2f} €")
```

#### Plot actual vs predicted

```python
plt.figure(figsize=(10, 5))
plt.plot(test["ds"], test["y"], label="Actual")
plt.plot(pred.index, pred["yhat"], label="Forecast")
plt.title("Test Set: Actual vs Forecasted Prices")
plt.xlabel("Date")
plt.ylabel("Price (€)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Octoberfest holiday effects

#### Define Octoberfest as a custom holiday

```python
import pandas as pd
oktoberfest = pd.DataFrame({
    'holiday': 'Oktoberfest',
    'ds': pd.to_datetime([
        '2016-09-17','2017-09-16','2018-09-22', '2019-09-21', '2022-09-17', '2023-09-16', '2024-09-21'
    ]),
    'lower_window': 0,
    'upper_window': 16  # lasts around 16–18 days
})
```

#### Combine into one DataFrame

```python
holidays = pd.concat([oktoberfest])
```

#### Pass holidays to the model

```python
from prophet import Prophet
model = Prophet(holidays=holidays)
```

#### Fit and forecast as usual

```python
model.fit(ts)

future = model.make_future_dataframe(periods=12, freq='W')
forecast = model.predict(future)
```

#### Plot with holidays

```python
fig = model.plot(forecast)
fig.suptitle("Airbnb Price Forecast with Holidays", fontsize=16)
```

```python
fig2 = model.plot_components(forecast)
```

```python
forecast_subset = forecast[["ds", "yhat", "holidays", "yhat_lower", "yhat_upper"]]
forecast_subset[forecast_subset["holidays"].abs() > 0].sort_values("ds")
```

# Forecasting by neighbourhood (Obergiesing)

```python
import pandas as pd
```

#### Load cleaned data

```python
df = pd.read_csv("../data/listings_cleaned.csv", parse_dates=["last_review"])
```

#### Drop missing review dates

```python
df = df.dropna(subset=["last_review"])
df = df[df["last_review"] >= "2024-01-01"]
```

#### Top 5 most common neighborhoods by listing count

```python
top_neigh = df["neighbourhood"].value_counts().nlargest(2).index.tolist()
top_neigh = ['Obergiesing']
print("Top neighborhoods:", top_neigh)
```

```python
neigh_ts = {}

for neigh in top_neigh:
    df_n = df[df["neighbourhood"] == neigh]
    ts = df_n.set_index("last_review").resample("W")["price"].mean().dropna()
    
    # Only keep if there are enough data points (e.g. 50+ weeks)
    if len(ts) >= 1:
        temp = ts.reset_index().rename(columns={"last_review": "ds", "price": "y"})
        neigh_ts[neigh] = temp
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

for neigh, data in neigh_ts.items():
    print(f"\nForecasting for: {neigh}")
    
    # Add holidays if you defined them earlier
    model = Prophet(holidays=holidays)
    model.fit(data)
    
    future = model.make_future_dataframe(periods=12, freq="W")
    forecast = model.predict(future)
    
    # Plot
    fig = model.plot(forecast)
    fig.suptitle(f"Forecast for {neigh}", fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Average Price (€)")
    plt.show()
    fig2 = model.plot_components(forecast)
```
